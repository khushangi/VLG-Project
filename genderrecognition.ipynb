```python
import os
import joblib
from sklearn.mixture import GaussianMixture
import librosa
import numpy as np

def extract_features(audio_path, num_mfcc=13, n_fft=2048, hop_length=512):
    try:
        # Load audio file using the audioread backend
        y, sr = librosa.load(audio_path, sr=None, res_type='kaiser_best')

        # Extract MFCCs
        mfccs = librosa.feature.mfcc(y=y, sr=sr, n_mfcc=num_mfcc, n_fft=n_fft, hop_length=hop_length)

        # Transpose the matrix to have time along the rows and features along the columns
        mfccs = mfccs.T

        return mfccs

    except Exception as e:
        print(f"Error extracting features from {audio_path}: {e}")
        return None

def load_gmm_models(model_folder):
    models = {}
    genders = ["females", "males"]

    for gender in genders:
        model_path = os.path.join(model_folder, f"{gender}_gmm_model.pkl")
        if os.path.exists(model_path):
            model = joblib.load(model_path)
            models[gender] = model
        else:
            print(f"Model for {gender} not found at {model_path}")

    return models

def test_system(models, audio_path):
    extracted_features = extract_features(audio_path)

    if extracted_features is not None:
        scores = {}
        for gender, model in models.items():
            log_likelihood = model.score(extracted_features)
            scores[gender] = log_likelihood

        predicted_gender = max(scores, key=scores.get)
        print(f"Predicted gender for {audio_path}: {predicted_gender}")
    else:
        print(f"Feature extraction failed for {audio_path}")

if __name__ == "__main__":
    dataset_path = r"C:\Users\91820\Downloads\SLRVLG\TestingData"
    model_folder = r"C:\Users\91820\Downloads\SLRVLG\GMMModels"

    # Load GMM models
    gmm_models = load_gmm_models(model_folder)

    # Test the system on each audio file in the testing dataset
    for filename in os.listdir(dataset_path):
        audio_path = os.path.join(dataset_path, filename)
        test_system(gmm_models, audio_path)
```
